```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
iris = load_iris()
```

```python
X, y = iris.data, iris.target
```

### Decision Tree

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(X, y)
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(model, out_file=None)
```

```python
graph = graphviz.Source(dot_data)
```

```python
graph
```

여기까지가 이전 시간에 했던 Decision Tree

### Random Forest

```python
# Random Forest는 앙상블 기법 중 하나이므로 sklearn의 ensemble 패키지 안에 있다.
# decision tree와 사용 방법이 거의 비슷하다.

from sklearn.ensemble import RandomForestClassifier
```

```python
# __init__을 보면 여러가지 요소들이 보인다. 그 중 유심히 살펴봐야 할 것들은
# max_depth의 경우 트리의 깊이를 정하는 매개변수이다.
# n_estimators는 트리를 몇개 만들것이냐, 기본값은 100개이다.
# min_samples_split은 샘플을 최소 몇개로 쪼갤것인지이다.

rf_model = RandomForestClassifier(n_estimators=10, max_depth=4, random_state=0)
```

```python
rf_model = rf_model.fit(X, y)
```

```python
rf_model.predict([X[0], X[50], X[130]])
```
